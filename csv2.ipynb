```python
import cv2
import numpy as np

def obtener_propiedades_nubes(imagen_path):
    # Cargar la imagen en escala de grises
    imagen = cv2.imread(imagen_path, 0)

    # Obtener las dimensiones de la imagen
    alto, ancho = imagen.shape[:2]
    centro_x = ancho // 2
    centro_y = alto // 2

    # Calcular la distancia desde el centro de la imagen hacia los bordes
    distancia_max = int(min(centro_x, centro_y))

    # Calcular los ángulos en sentido antihorario desde el eje x positivo hacia el eje y positivo
    angulos = np.arange(0, 2 * np.pi, np.pi / 180)

    # Lista para almacenar las propiedades de las nubes
    propiedades_nubes = []

    # Dibujar los ejes con contorno rojo
    imagen_color = cv2.cvtColor(imagen, cv2.COLOR_GRAY2BGR)
    cv2.line(imagen_color, (centro_x, 0), (centro_x, alto), (0, 0, 255), 2)
    cv2.line(imagen_color, (0, centro_y), (ancho, centro_y), (0, 0, 255), 2)

    # Detectar los contornos de los objetos desde el centro hacia afuera
    for distancia in range(distancia_max, 0, -1):
        # Calcular el radio para el círculo actual
        radio = int(distancia * 0.254)

        # Detectar los contornos en el círculo actual
        contornos, _ = cv2.findContours(imagen, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Calcular las propiedades para cada contorno de nube
        for i, contorno in enumerate(contornos):
            # Calcular el área, perímetro, centroides y orientación del contorno
            area = cv2.contourArea(contorno)
            perimetro = cv2.arcLength(contorno, True)
            momentos = cv2.moments(contorno)
            cX = momentos['m10'] / momentos['m00']
            cY = momentos['m01'] / momentos['m00']
            (x, y, ancho_contorno, alto_contorno) = cv2.boundingRect(contorno)
            relacion_ancho = ancho_contorno / alto_contorno
            (latitud, longitud) = obtener_latitud_longitud(cX, cY)  # Función para calcular la latitud y longitud a partir de cX y cY
            angulo = obtener_orientacion(contorno)  # Función para calcular la orientación del contorno
            intensidad = obtener_intensidad(contorno)  # Función para calcular la intensidad de la nube

            # Dibujar el contorno de la nube en negro en la imagen
            cv2.drawContours(imagen_color, [contorno], 0, (0, 0, 0), 2)

            # Dibujar el contorno del círculo actual en contorno rojo
            cv2.circle(imagen_color, (centro_x, centro_y), radio, (0, 0, 255), 2)

            # Crear un diccionario con las propiedades de la nube actual
            propiedades = {
                "id": i + 1,
                "tiempo": 0,
                "contorno_x": x * 0.254,
                "contorno_y": y * 0.254,
                "centroide_x": cX * 0.254,
                "centroide_y": cY * 0.254,
                "latitud": latitud,
                "longitud": longitud,
                "ancho": ancho_contorno,
                "relacion_ancho": relacion_ancho,
                "area": area * 0.254 * 0.254,
                "perimetro": perimetro * 0.254,
                "orientacion": angulo,
                "intensidad": intensidad
            }

            # Agregar las propiedades a la lista
            propiedades_nubes.append(propiedades)

    # Mostrar la imagen con los contornos y ejes
    cv2.imshow('Imagen con contornos y ejes', imagen_color)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return propiedades_nubes

# Función para calcular la latitud y longitud a partir de cX y cY
def obtener_latitud_longitud(cX, cY):
    # Cálculos para obtener la latitud y longitud
    latitud = 42.123  # Ejemplo: valor constante para la latitud
    longitud = -73.456  # Ejemplo: valor constante para la longitud

    return latitud, longitud

# Función para calcular la orientación del contorno
def obtener_orientacion(contorno):
    # Cálculo de la orientación del contorno
    angulo = 45  # Ejemplo: valor constante para la orientación

    return angulo

# Función para calcular la intensidad de la nube
def obtener_intensidad(contorno):
    # Cálculo de la intensidad de la nube
    intensidad = 150  # Ejemplo: valor constante para la intensidad

    return intensidad

# Ruta de la imagen geoespacial
imagen_path = '/home/sebastian/Documentos/analisisNETCDF/img.png'

# Obtener las propiedades de las nubes
propiedades_nubes = obtener_propiedades_nubes(imagen_path)

# Imprimir las propiedades obtenidas
for propiedades in propiedades_nubes:
    print(propiedades)
```
