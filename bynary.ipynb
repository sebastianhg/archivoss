```python
import cv2

def binarizar_imagen(imagen):
    """
    Binariza una imagen utilizando un umbral fijo.

    Args:
      imagen: La imagen a binarizar.

    Returns:
      La imagen binarizada.
    """

    # Convertir la imagen a escala de grises
    imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

    # Definir el umbral
    umbral = 128  # Puedes ajustar este valor según tus necesidades

    # Aplicar la binarización
    _, imagen_binarizada = cv2.threshold(imagen_gris, umbral, 255, cv2.THRESH_BINARY)

    return imagen_binarizada

# Leer la imagen
imagen = cv2.imread("imagen_coloreada_contornos_cafe.png")

# Binarizar la imagen
imagen_binarizada = binarizar_imagen(imagen)

# Mostrar la imagen binarizada
cv2.imshow("Imagen Binarizada", imagen_binarizada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
