```python
import cv2
import numpy as np
from skimage import measure
import os

# Ruta de la imagen
ruta_imagen = '/home/sebastian/Documentos/analisisNETCDF/objetos/2022-01-19_objeto_1.png'

# Lee la imagen en escala de grises
imagen = cv2.imread(ruta_imagen, 0)

# Aplica umbralización para binarizar la imagen
_, binarizada = cv2.threshold(imagen, 128, 255, cv2.THRESH_BINARY)

# Invierte los colores de la imagen binarizada
binarizada_invertida = cv2.bitwise_not(binarizada)

# Encuentra los componentes conectados en la imagen binarizada invertida
componentes = measure.label(binarizada_invertida, connectivity=2)

# Encuentra el centroide del componente más grande (número en el centro)
propiedades = measure.regionprops(componentes)
centroide = propiedades[0].centroid

# Crea una imagen en blanco y negro para mostrar los bits
imagen_bits = np.zeros((binarizada.shape[0], binarizada.shape[1]), dtype=np.uint8)

# Asigna los valores binarizados a la imagen de bits
imagen_bits[binarizada_invertida == 255] = 1

# Dibuja un círculo en el centroide del número
imagen_visualizacion = cv2.cvtColor(imagen_bits * 255, cv2.COLOR_GRAY2BGR)
cv2.circle(imagen_visualizacion, (int(centroide[1]), int(centroide[0])), 5, (0, 255, 0), -1)

# Guarda la imagen binarizada en la carpeta "binarizado" en la ubicación actual
carpeta_salida = "binarizado"
nombre_imagen = "imagen_binarizada.png"
ruta_salida = os.path.join(carpeta_salida, nombre_imagen)
cv2.imwrite(ruta_salida, imagen_visualizacion)

# Muestra la imagen de bits con el número y el centroide resaltados
cv2.imshow('Imagen binaria invertida', imagen_visualizacion)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Ruta de la imagen
ruta_imagen = '/home/sebastian/Documentos/analisisNETCDF/objetos/2022-01-19_objeto_1.png'

# Lee la imagen
imagen = cv2.imread(ruta_imagen, 0)

# Aplica un umbral para binarizar la imagen
_, binarizada = cv2.threshold(imagen, 128, 255, cv2.THRESH_BINARY)

# Cambia el tamaño de la imagen binarizada a 1000x600
binarizada_resized = cv2.resize(binarizada, (1000, 600))

# Encuentra los componentes conectados en la imagen binarizada
componentes = measure.label(binarizada_resized, connectivity=2)

# Encuentra el centroide del componente más grande (número en el centro)
propiedades = measure.regionprops(componentes)
centroide = propiedades[0].centroid

# Crea una imagen en blanco para mostrar los bits
imagen_bits = np.zeros((600, 1000, 3), dtype=np.uint8)

# Asigna los valores binarizados a los canales R, G y B de la imagen de bits
imagen_bits[:, :, 2] = binarizada_resized
imagen_bits[:, :, 1] = binarizada_resized
imagen_bits[:, :, 0] = binarizada_resized

# Dibuja un círculo en el centroide del número
cv2.circle(imagen_bits, (int(centroide[1]), int(centroide[0])), 5, (0, 255, 0), -1)

# Muestra la imagen de bits con el número y el centroide resaltados
cv2.imshow('Bits de la imagen', imagen_bits)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
