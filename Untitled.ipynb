```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_coordinates_from_csv(csv_file):
    # Paso 2: Leer los datos desde el archivo CSV
    df = pd.read_csv(csv_file)

    # Paso 3: Graficar las coordenadas sin animacin
    plt.figure(figsize=(10, 8))
    plt.scatter(df['longitude'], df['latitude'], c=df['ivt'], cmap='viridis', s=70*df['ivt'], marker='o', alpha=0.8)  # Marcador reducido en un 30%
    plt.colorbar(label='IVT')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.title('Grfico de objeto')
    plt.grid(True)

    # Ajustar los lmites de los ejes para una vista ms alejada
    plt.xlim(df['longitude'].min() - 1, df['longitude'].max() + 1)
    plt.ylim(df['latitude'].min() - 1, df['latitude'].max() + 1)

    plt.show()

# Ruta al archivo CSV con las coordenadas
archivo_csv = '/home/sebastian/Documentos/analisisNETCDF/conjuntos/conjunto_1.csv'
plot_coordinates_from_csv(archivo_csv)

```
