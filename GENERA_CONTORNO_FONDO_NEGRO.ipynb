```python
import cv2
import numpy as np

def visualizar_objetos(imagen):
    """
    Visualiza los objetos azules y grandes en una imagen con diferentes colores y contornos.

    Args:
      imagen: La imagen a visualizar.

    Returns:
      La imagen con los objetos visualizados.
    """

    # Convertir la imagen de BGR a HSV
    imagen_hsv = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)

    # Definir los rangos de color azul en el espacio de color HSV
    lower_blue = np.array([100, 50, 50])  # Límite inferior del azul
    upper_blue = np.array([130, 255, 255])  # Límite superior del azul

    # Crear una máscara para los píxeles azules en la imagen
    mask_blue = cv2.inRange(imagen_hsv, lower_blue, upper_blue)

    # Encontrar contornos en la máscara azul
    contours, _ = cv2.findContours(mask_blue, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Crear una imagen en blanco del mismo tamaño que la original
    h, w = mask_blue.shape[:2]
    objetos_coloreados = np.zeros((h, w, 3), dtype=np.uint8)

    # Dibujar contornos y llenar objetos grandes con diferentes colores
    for i, contour in enumerate(contours):
        # Calcular el área del objeto
        area = cv2.contourArea(contour)

        # Definir un umbral para identificar objetos grandes
        umbral_area = 5000  # Puedes ajustar este valor según tus necesidades

        # Si el área del objeto es mayor que el umbral, dibujar el contorno y llenar el objeto
        if area > umbral_area:
            # Dibujar el contorno del objeto
            cv2.drawContours(objetos_coloreados, [contour], -1, (0, 255, 0), 2)

            # Crear una máscara para el objeto
            mask = np.zeros_like(mask_blue)
            cv2.drawContours(mask, [contour], 0, 255, thickness=cv2.FILLED)

            # Elegir un color aleatorio
            color = list(np.random.randint(0, 256, size=(3,), dtype=np.uint8))

            # Aplicar el color a la región del objeto en la imagen coloreada
            objetos_coloreados[mask > 0] = color

    return objetos_coloreados

# Leer la imagen
imagen = cv2.imread("objetos_azules.png")

# Visualizar los objetos con diferentes colores y contornos (eliminando objetos pequeños)
imagen_coloreada = visualizar_objetos(imagen)

# Guardar la imagen final
cv2.imwrite("imagen_coloreada_contornos.png", imagen_coloreada)

print("Imagen con objetos azules grandes coloreados y contornos guardada como 'imagen_coloreada_contornos.png'")import cv2
import numpy as np
from skimage.morphology import skeletonize
import os

def obtener_nubes_blancas(imagen):
  """Obtiene las nubes blancas de la imagen."""

  hsv = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)
  blanco_bajo = np.array([0, 0, 200])
  blanco_alto = np.array([255, 30, 255])
  mascara_blanca = cv2.inRange(hsv, blanco_bajo, blanco_alto)
  return mascara_blanca

def binarizar_imagen(imagen, tamano_objeto):
  """Binaria la imagen de las nubes."""

  # Umbral de la imagen
  _, imagen_binaria = cv2.threshold(imagen, 1, 255, cv2.THRESH_BINARY)

  # Dilata la imagen para rellenar los huecos
  imagen_binaria = cv2.dilate(imagen_binaria, np.ones((tamano_objeto, tamano_objeto), np.uint8))

  return imagen_binaria

def esqueletonizar_imagen(imagen_binaria):
  """Obtiene el esqueleto de la imagen."""

  return skeletonize(imagen_binaria)

# Ruta de la carpeta que contiene las imágenes
ruta_carpeta = '/home/sebastian/Documentos/analisisNETCDF/datosGraficados/2022-01_binarizadoAxis/'

# Ruta de la carpeta donde se guardarán las imágenes binarizadas
carpeta_salida = '/home/sebastian/Documentos/analisisNETCDF/datosGraficados/2022/2022-01_axiss'
os.makedirs(carpeta_salida, exist_ok=True)

# Obtener la lista de archivos en la carpeta de imágenes
archivos_en_carpeta = os.listdir(ruta_carpeta)

for archivo in archivos_en_carpeta:
  # Filtrar solo los archivos de imagen (por ejemplo, con extensión .jpg o .png)
  if archivo.lower().endswith(('.jpg', '.png', '.jpeg')):
    # Construir la ruta completa de la imagen
    ruta_imagen = os.path.join(ruta_carpeta, archivo)

    # Cargar la imagen
    imagen = cv2.imread(ruta_imagen)

    if imagen is None:
      print(f"No se pudo cargar la imagen {ruta_imagen}, asegúrate de que la ruta sea correcta.")
      continue

    # Realizar el procesamiento de la imagen como en tu código original
    mascara_nubes_blancas = obtener_nubes_blancas(imagen)
    nubes_binarias = binarizar_imagen(imagen, 10)
    nubes_esqueleto = esqueletonizar_imagen(nubes_binarias)

    # Guardar la imagen binaria y su esqueleto en la carpeta de salida
    nombre_binaria = f'nubes_binarias_{archivo}'
    nombre_esqueleto = f'nubes_esqueleto_{archivo}'
    cv2.imwrite(os.path.join(carpeta_salida, nombre_binaria), nubes_binarias)
    cv2.imwrite(os.path.join(carpeta_salida, nombre_esqueleto), nubes_esqueleto)

    print(f"Procesada la imagen: {archivo}")

print("Procesamiento completado.")
```
