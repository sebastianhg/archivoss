```python
import cv2
import numpy as np
import os

def visualizar_objetos_azules_claros(ruta_imagen):
    """
    Detecta los objetos azules claros en una imagen y les pone un contorno.

    Args:
      ruta_imagen: La ruta de la imagen a procesar.

    Returns:
      La imagen con los objetos azules contorneados.
    """

    # Leer la imagen
    imagen = cv2.imread(ruta_imagen)

    if imagen is None:
        print(f"Error: No se pudo abrir la imagen '{ruta_imagen}'.")
        return None

    # Convertir la imagen de BGR a HSV (Espacio de Color Hue-Saturation-Value)
    imagen_hsv = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)

    # Definir el rango de color azul claro en HSV
    lower_blue = np.array([90, 50, 150])  # Valores ajustados para tonos más claros
    upper_blue = np.array([130, 255, 255])

    # Crear una máscara para los píxeles azules claros en la imagen
    mask = cv2.inRange(imagen_hsv, lower_blue, upper_blue)

    # Encontrar contornos en la máscara
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Dibujar contornos alrededor de los objetos azules claros
    imagen_contorneada = imagen.copy()
    cv2.drawContours(imagen_contorneada, contours, -1, (0, 255, 0), 2)

    return imagen_contorneada

# Carpeta de entrada y salida
carpeta_entrada = "time6"
carpeta_salida = "objetos_contornos_azul_claro"

# Crear la carpeta de salida si no existe
if not os.path.exists(carpeta_salida):
    os.makedirs(carpeta_salida)

# Listar los archivos en la carpeta de entrada
archivos = os.listdir(carpeta_entrada)

# Procesar cada imagen en la carpeta de entrada
for archivo in archivos:
    # Construir la ruta completa de la imagen de entrada
    ruta_imagen_entrada = os.path.join(carpeta_entrada, archivo)

    # Procesar la imagen
    imagen_contornos = visualizar_objetos_azules_claros(ruta_imagen_entrada)

    if imagen_contornos is not None:
        # Construir la ruta de la imagen de salida
        ruta_imagen_salida = os.path.join(carpeta_salida, f"contornos_azul_claro_{archivo}")

        # Guardar la imagen procesada en la carpeta de salida
        cv2.imwrite(ruta_imagen_salida, imagen_contornos)
        print(f"Imagen '{archivo}' procesada y guardada como '{ruta_imagen_salida}'.")
    else:
        print(f"Error al procesar la imagen '{archivo}'.")
```
